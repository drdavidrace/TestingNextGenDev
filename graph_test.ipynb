This code presents a simple graph of $y = x^2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3.0, 3.0,.01)
y = x**2.0

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title(r'$y = x^2$')
plt.grid(True)
plt.show()
```
